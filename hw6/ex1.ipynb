/* 1. Создайте таблицу users_old, аналогичную таблице users. Создайте процедуру, с помощью которой
  можно переместить любого (одного) пользователя из таблицы users в таблицу users_old.
   (использование транзакции с выбором commit или rollback – обязательно).
*/

-- Создадим структурную копию таблицы users.

drop table if exists users_old;
create table users_old like users;

select * from users_old;

-- Создадим процедуру с нужным функционалом используя разделитель. Предусмотрим отчет об ошибках.

drop procedure if exists sp_transfer;

delimiter //
create procedure sp_transfer(in user_num int, out trans_res varchar(100))
begin
	
	declare `_rollback` bit default b'0';
	declare code varchar(100);
	declare error_str varchar(100);

	declare continue handler  for sqlexception
	begin set `_rollback` = b'1';
	get stacked diagnostics condition 1
		code = returned_sqlstate, error_str = message_text;
	end;
		
	start transaction;

	insert into users_old select * from users where id = user_num;
	delete from users where id = user_num;

	if `_rollback` then
		set trans_res = concat('Ошибка: ', code, 'Описание ошибки: ', error_str);
	rollback;
	else set trans_res = 'все норм';
		commit;
	end if;

end//
delimiter ;

call sp_transfer(1, @trans_res);
select @trans_res;

-- Проверим что созданная процедура работает корректно. Данный пользовтаеля с заданным id должны появиться
-- в users_old и удалиться из users

/*Создайте хранимую функцию hello(), которая будет возвращать приветствие, в зависимости от текущего времени суток.
С 6:00 до 12:00 функция должна возвращать фразу "Доброе утро",
с 12:00 до 18:00 функция должна возвращать фразу "Добрый день",
с 18:00 до 00:00 — "Добрый вечер", с 00:00 до 6:00 — "Доброй ночи".
*/

drop function if exists hello

delimiter //
create function hello() returns varchar(20) deterministic
begin
	return case
      	when current_time() between '06:00:00' and '11:59:59' then "Доброе утро"	
		when current_time() between '12:00:00' and '17:59:59' then "Добрый день"	
		when current_time() between '18:00:00' and '23:59:59' then "Добрый вечер"	
		else "Доброй ночи"		
	end;
end //

delimiter ;

select hello();
