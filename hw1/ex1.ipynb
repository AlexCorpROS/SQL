CREATE SCHEMA `hw1` ;

CREATE TABLE `hw1`.`mobile_phones` (
  `id` INT NOT NULL AUTO_INCREMENT,
  `product_name` VARCHAR(45) NOT NULL,
  `manufacturer` VARCHAR(45) NOT NULL,
  `product_count` INT NOT NULL,
  `price` DECIMAL NULL,
  PRIMARY KEY (`id`));
